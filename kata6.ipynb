```python
# creamos la lista de planetas
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
# para obtener la longitud de la lista usamos el metodo len
planets2 = (len(planets))
print(planets2)

# Agregamos a pluto a la lista con el metodo .append('pluto')
planets.append('pluto')
planets2 = (len(planets))
print(planets2)
```

# ejercicio 2

```python
# creamos la lista de planetas
planetas = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
nombre_planeta = input("Ingre el nombre de un planeta 'con la primer letra en mayuscula'")
# Buscamos el planeta en la lista con el metodo .index()
planeta = planetas.index(nombre_planeta)
print(planeta)
# mostrar los planetas mas cercanos al sol 
print('Los planetas mas cercanos al sol antes que '+ nombre_planeta)
print(planetas[0:planeta])
# mostrar el planeta mas lejano al sol
print('Los planetas mas lejanos al sol que '+ nombre_planeta)
print(planetas[planeta+1:])
```
