```python
# creamos un diccionario llamado planet
planet = {
    'name':'Mars',
    'moons':2
}
# mostramos el nombre y numero de lunas con el metodo get y corchetes ([])
print(planet.get('name'))
print(planet['moons'])
# Agregamos una nueva clave 
planet['circunferencia (km)'] = {
    'polar' : 6752,
    'equatorial' : 6792
}
# imprimimos el nombre del planeta con su circunferencia polar
print (f'{planet["name"]} su circunferencia polar es: {planet["circunferencia (km)"]["polar"]}')
```

```python
planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1

}
moons= planet_moons.values()
planets = len(planet_moons.keys())
total_moons = 0
for moon in moons :
    total_moons = total_moons + moons
average = total_moons / planets
print(average)
```
