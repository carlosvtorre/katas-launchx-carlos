```python
# biblotecas math
## ceil: es para redondear un numero hacia arriba
## floor: es para redondear un numero hacia abajo
from math import ceil,floor
redondeo_arriba= ceil(12.5)
print(redondeo_arriba)
redondeo_abajo= floor(12.5)
print(redondeo_abajo)
```

```python
# para el valor absoluto se usa el abs es decir para comverrtir el resultado negativo en positivo
a=39-16
b=16-39
print(abs(a))
print(abs(b))
# La funcion (round) redonde el numero decimal al entero mas cercano
print(round(14.5))
```
